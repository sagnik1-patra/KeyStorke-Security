```python
import pandas as pd
import os
import time
import keyboard
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import to_categorical

# ✅ Step 1: Load data
file_paths = [
    r"C:\Users\sagni\Downloads\KeyStrock Security\Annan_Sadr.csv",
    r"C:\Users\sagni\Downloads\KeyStrock Security\Rahi_Pandey.csv",
    r"C:\Users\sagni\Downloads\KeyStrock Security\Sagnik_Patra.csv"
]

dataframes = []
for path in file_paths:
    df = pd.read_csv(path)
    name = os.path.splitext(os.path.basename(path))[0]
    df['user'] = name
    dataframes.append(df)

df_all = pd.concat(dataframes, ignore_index=True)

# ✅ Step 2: Prepare features and labels
X = df_all.drop(columns=["user"])
y = df_all["user"]

label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
y_cat = to_categorical(y_encoded)

X_train, X_test, y_train, y_test = train_test_split(X, y_cat, test_size=0.2, random_state=42)

# ✅ Step 3: Train model
model = Sequential([
    Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    Dense(64, activation='relu'),
    Dense(len(label_encoder.classes_), activation='softmax')
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=50, batch_size=8, validation_split=0.2)

# ✅ Step 4: Save model
model_path = r"C:\Users\sagni\Downloads\KeyStrock Security\keystroke_user_model.h5"
model.save(model_path)

print(f"✅ Model saved to {model_path}")
```

```python
import os
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras.utils import to_categorical

# 📂 Paths
data_dir = r"C:\Users\sagni\Downloads\KeyStrock Security\Data"
model_path = os.path.join(data_dir, "keystroke_user_model.h5")

# 📥 Load and prepare dataset
data = []
labels = []

for file in os.listdir(data_dir):
    if file.endswith(".csv") and file != "test.csv":
        file_path = os.path.join(data_dir, file)
        try:
            df = pd.read_csv(file_path)
            if df.shape[1] == 16:  # ✔️ Use only files with 16 features
                df = df.select_dtypes(include=[np.number])  # ✔️ Ensure numeric
                data.append(df)
                labels.extend([file.replace(".csv", "")] * len(df))
            else:
                print(f"⚠️ Skipping {file}: Expected 16 features, got {df.shape[1]}")
        except Exception as e:
            print(f"❌ Failed to load {file}: {e}")

if not data:
    raise ValueError("No valid CSV files found!")

X = pd.concat(data, ignore_index=True).astype("float32")
y = np.array(labels)

# 🔠 Encode labels
le = LabelEncoder()
y_enc = le.fit_transform(y)
y_cat = to_categorical(y_enc)

# 🧪 Split data
X_train, X_test, y_train, y_test = train_test_split(X, y_cat, test_size=0.2, random_state=42)

# 🧠 Build model
model = Sequential([
    Input(shape=(16,)),
    Dense(64, activation='relu'),
    Dense(32, activation='relu'),
    Dense(y_cat.shape[1], activation='softmax')
])
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# 🚂 Train model
model.fit(X_train, y_train, epochs=30, batch_size=8, validation_data=(X_test, y_test), verbose=1)

# 💾 Save model
model.save(model_path)

# 🔍 Prediction from test.csv
test_file = os.path.join(data_dir, "test.csv")
if os.path.exists(test_file):
    test_df = pd.read_csv(test_file).astype("float32")
    if test_df.shape[1] != 16:
        raise ValueError("Test file must have 16 features!")
    
    model = load_model(model_path)
    pred = model.predict(test_df)
    pred_class = np.argmax(pred, axis=1)
    pred_label = le.inverse_transform(pred_class)
    print("🧑 Predicted user:", pred_label[0])
else:
    print("❌ test.csv not found")
```

```python
import os
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense, Dropout, Input
from tensorflow.keras.optimizers import Adam

# 📂 Path to folder with all keystroke CSVs
folder_path = r"C:\Users\sagni\Downloads\KeyStrock Security\Data"

# 📥 Load all datasets
data = []
labels = []

for file in os.listdir(folder_path):
    if file.endswith('.csv'):
        file_path = os.path.join(folder_path, file)
        try:
            df = pd.read_csv(file_path)
            if df.shape[1] == 16:  # Expecting 16 features
                data.append(df)
                labels.extend([file.replace('.csv', '')] * len(df))
            else:
                print(f"⚠️ Skipping {file}: Expected 16 features, got {df.shape[1]}")
        except Exception as e:
            print(f"❌ Failed to load {file}: {e}")

if not data:
    raise ValueError("❌ No valid CSV files found!")

# 🧪 Prepare features and labels
X = pd.concat(data, ignore_index=True)
y = np.array(labels)

# 🔢 Normalize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 🔤 Encode labels
encoder = LabelEncoder()
y_encoded = encoder.fit_transform(y)
y_cat = to_categorical(y_encoded)

# 🧠 Build deeper neural network
model = Sequential([
    Input(shape=(16,)),
    Dense(128, activation='relu'),
    Dropout(0.3),
    Dense(64, activation='relu'),
    Dropout(0.3),
    Dense(32, activation='relu'),
    Dense(y_cat.shape[1], activation='softmax')
])

# 🛠️ Compile model
model.compile(optimizer=Adam(learning_rate=0.001),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# 🏋️‍♂️ Train model
model.fit(X_scaled, y_cat, epochs=100, batch_size=16, validation_split=0.2)

# 💾 Save model and scaler
model_path = os.path.join(folder_path, "keystroke_user_model.h5")
model.save(model_path)
print(f"✅ Model saved at {model_path}")
```
